```python
import pandas as pd
import numpy as np
import requests

import psycopg2
import os
```

```python
res2 = requests.get("http://isin.twse.com.tw/isin/C_public.jsp?strMode=2")
res4 = requests.get("http://isin.twse.com.tw/isin/C_public.jsp?strMode=4")

df2 = pd.read_html(res2.text)[0]
df4 = pd.read_html(res4.text)[0]

df2.columns = df2.iloc[0]
df4.columns = df4.iloc[0]

df2 = df2.iloc[1:]
df4 = df4.iloc[1:]

df2 = df2[df2['產業別'].notna()]
df2 = df2[df2['CFICode']=='ESVUFR']

df4 = df4[df4['產業別'].notna()]
df4 = df4[df4['CFICode']=='ESVUFR']

df2_arr = np.array(df2['有價證券代號及名稱'])
df4_arr = np.array(df4['有價證券代號及名稱'])
```

```python
TSE_Size = df2_arr.size
OTC_Size = df4_arr.size

TSE_List = list(range(TSE_Size))
OTC_List = list(range(OTC_Size))

TSE_DF = pd.DataFrame(columns=['TSE_Number', 'TSE_Name'],index=TSE_List)
OTC_DF = pd.DataFrame(columns=['OTC_Number', 'OTC_Name'],index=OTC_List)
```

```python
for i in range(df2_arr.size):
    TSE_DF.iloc[i][0] = df2_arr[i][:4]
    TSE_DF.iloc[i][1] = df2_arr[i][5:8]
    
for i in range(df4_arr.size):
    OTC_DF.iloc[i][0] = df4_arr[i][:4]
    OTC_DF.iloc[i][1] = df4_arr[i][5:8]
```

```python
TSE_DF.to_csv(filepath + r"\TSE_DF.csv")
OTC_DF.to_csv(filepath + r"\OTC_DF.csv")
```
